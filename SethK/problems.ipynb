## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random
import numpy as np

def dice_roll(simulation_amount = 1000): # Create a function that stimulates rolling the dice and have this repeat 1000 times. 
    count = 0                               # start the count at 0

    for i in range(simulation_amount):
        dice_one = random.randint(1, 6)   # The randint is choosing a random number from 1 to 6
        dice_two = random.randint(1, 6)
        
        score_sum = dice_one + dice_two
        
        if (score_sum % 2 == 0) or (score_sum > 7):
            count += 1
            # count+= 1 is same as count = count + 1. 
            #This our way of keeping track of the number of times that the sum was either greater than 7 or an even number.

    return count / simulation_amount
    
print(f"The probability of rolling an even number or greater than 7 is: {np.round(dice_roll() * 100, 2)}%")

```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import numpy as np
import random
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"
        
print(ball_box)
            
def grab_balls(sim_amount = 1000):
    #Now we intialize the counts to 0
    countOne = 0
    countTwo = 0
    
    #Create a for loop
    for i in range(sim_amount):
        #Create an empty list to store the colors
        colors = []

        #simluate taking 5 balls from the box
        for i in range(5):
            colors.append(ball_box[np.random.randint(0, 59)])
        
        # The list would have to be converted to a numpy array 
        colors = np.array(colors)

        #Next we find the sums of each ball being picked
        white_chosen = sum(colors == 'White')
        red_chosen = sum(colors == 'Red')
        green_chosen = sum(colors == 'Green')

        #Now what we have to do is decide if the count should be added to if the certain events happen
        if (white_chosen == 3) and (red_chosen == 2):
            countOne += 1
        
        if (white_chosen == 5) or (red_chosen == 5) or (green_chosen == 5):
            countTwo += 1
    #Now what we return is the counts divided by the amount of times the actions was performed. In this case, 1000 times.
    return countOne / sim_amount, countTwo / sim_amount

#We call the function
probability = grab_balls(1000)

#Now we print our result
print(f"The probability of picking 3 white balls and 2 red balls is:{np.round(probability[0] * 100, 2)}%")
print(f"The probability of all the same color balls is: {np.round(probability[1] * 100, 2)}%")
```
