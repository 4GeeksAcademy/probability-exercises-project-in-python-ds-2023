## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np

def roll_the_dice(roll_sum):
        if roll_sum > 7 and roll_sum % 2 == 0:
            return "AB" 
        if roll_sum > 7:
            return "A"
        if roll_sum % 2 == 0:
            return "B"
        
higher_than_7 = 0
even_result = 0

for sim in range(1000):
    array = []
    dice1 = np.random.randint(1,7)
    dice2 = np.random.randint(1,7)
    roll_sum = dice1 + dice2
    result = roll_the_dice(roll_sum)
    array.append(result)
    if result == "A":
            higher_than_7 +=1
    if result == "B":
            even_result +=1
    if result == "AB":
          higher_than_7 +=1
          even_result +=1
print("Number of times sum > 7:", higher_than_7)
print("Number of even results:", even_result)
print("Probability of sum > 7 or even result:", (higher_than_7 + even_result) / 1000.0)
    

    






```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
# Stores the amount of times 5 balls of the same color were grabbed
same_color = 0
# Stores the amount of times 3 white balls and 2 red balls were grabbed
three_white_two_red = 0
# Stores the color of the balls chosen per iteration of the experiment
ball_selections = []

# Runs the experiment 1000 times
for sim in range(1000):
    # Simulates grabbing 5 balls from the box
    def grab_five():
        for grab in range(5):
            # Determines the color of the ball chosen
            result = ball_box[np.random.randint(0, 59)]
            ball_selections.append(result)
    grab_five()
    white_balls = ball_selections.count('White')
    red_balls = ball_selections.count('Red')
    green_balls = ball_selections.count('Green')
    if (white_balls == 3) and (red_balls == 2):
        three_white_two_red += 1
    if (white_balls == 5) or (red_balls == 5) or (green_balls == 5):
        same_color += 1
    ball_selections.clear()    
print(f"The probability of 3 white and 2 red is: {np.round((three_white_two_red / 1000) * 100, 2)}%")
print(f"The probability of all the same color is: {np.round((same_color / 1000) * 100, 2)}%")    

```
